```python
# Write your code here
import math
import sympy
from sympy import symbols
from sympy.solvers import solve


# Add
def add():
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    result = a + b
    print(result)


# Subtract
def sub():
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    result = a - b
    print(result)


# Multiply
def mul(a, b):
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    result = a * b
    print(result)


# Divide
def div(a, b):
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))

    if b == 0:
        return "division by zero"

    result = a / b
    print(result)


# Detect prime numbers
def is_prime(number):
    result = True

    for test_number in range(2, number):
        if number % test_number == 0:
            return False

    return result


def prime():
    number = int(input("Enter the number to check: "))
    if is_prime(number):
        print("Prime")
    else:
        print("Not prime")


# Generate prime factors of a number
def prime_factors():
    num = int(input("Enter a number: "))
    factors = list()
    div = 2

    while num != 1:
        while num % div == 0:
            factors.append(div)
            num //= div
        div += 1

    print(factors)


# Simplify square roots
def sim_square(n):
    n = int(input("Enter a number: "))
    upper_limit = math.floor(math.sqrt(n)) + 1
    max_factor = 1
    other_factor = 1
    square_root = 1

    for maybe_factor in range(1, upper_limit):
        if n % (maybe_factor**2) == 0:
            max_factor = maybe_factor**2

    # Divide out the greatest square factor
    other_factor = n / max_factor

    # Output
    square_root = int(math.sqrt(max_factor))
    other_factor = int(other_factor)
    output = square_root * sympy.sqrt(other_factor)

    return output


# Solve for a variable
def solve_variable():
    x = symbols("x")
    eq = input("Enter an equation to solve for x: ")
    print("x = ", solve(eq, x)[0])


def power():
    n = int(input("Enter the base: "))
    ex = int(input("Enter the exponent: "))
    print(n**ex)

print('''Menu:
1. Add
2. Subtract
3. Multiply
4. Divide
5. Detect Prime Number
6. Generate Prime Factors
7. Simplify Square Root
8. Solve for a Variable
9. Power
10. Quit''')

# User input
cmd = input("Enter the number of your cmd: ")

if cmd == "1":
    add()
elif cmd == "2":
    sub()
elif cmd == "3":
    mul()
elif cmd == "4":
    div()
elif cmd == "5":
    prime()
elif cmd == "6":
    prime_factors()
elif cmd == "7":
    sim_square()
elif cmd == "8":
    solve_variable()
elif cmd == "9":
    power()
elif cmd == "10":
    print("Terminated")
else:
    print("Invalid")

# This step does not have test
```
